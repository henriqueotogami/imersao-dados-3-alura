# Aula 2: Estatísticas, visualização de dados e distribuições

## Substituindo o título da coluna de "droga" para "composto".

```python
mapa = {'droga':'composto'}
dados.rename(columns=mapa, inplace = True)
dados.head()
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]

cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

## Alterando o modelo de gráfico

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(10,8))
ax = sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 Compostos')
plt.show()
```

## Preparar os dados estatísticos para um histograma

```python
dados.loc[:, 'g-0':'g-771'].describe()
```

## Histograma da média

```python
dados.loc[:, 'g-0':'g-771'].describe().T['mean'].hist(bins=30)
```

## Cada "c" utilizado no histograma abaixo é um tipo ou linhagem celular. Pode ser oriunda de câncer ou saudáveis. Utilizado para saber se o composto afetou aquele tipo de célula, e ver como afetou.

```python
dados.loc[:, 'c-0':'c-99'].describe().T['mean'].hist(bins=50)
```
